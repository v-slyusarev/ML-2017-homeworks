** Задача 3 **
Докажите, что the k-means всегда сходится.

**Ответ.**
Пусть даны $N$ объектов и задано количество кластеров $m$. Заметим, что разбить $N$ объектов по $m$ кластерам можно $S(n,m)$ (число Стирлинга второго рода) способами - их конечное число. Тогда алгоритм проходит конечное число состояний, следовательно какое-то одно состояние принимается более одного раза. Более того, каждое состояние однозначно определяется предыдущим, так что после повторения какого-то состояния алгоритм строго зацикливается. Покажем, что длина такого цикла не превышает 1 действие. Отметим, что алгоритм минимизирует сумму расстояний от объектов до центров кластеров. Сумма $S = \sum_{i=1}^{m}\sum_{x_j \in C_i} \rho(x_j, \mu_i)$ является целевой функцией, которая не возрастает на каждом шаге алгоритма. Пусть $S[k]$ - значение целевой функции на $k$-ом шаге алгоритма. Выше показано, что алгоритм входит в цикл; пусть он начинается на шаге $n$ и имеет длину $l$, тогда в силу неубывания целевой функции $S[n] \le S[n+1] \le \ldots \le S[n+l]=S[n]$, откуда прямо следует, что все промежуточные значения $S[j],\ n < j < n+l$, в точности равны $S[n]$. Таким образом, целевая функция начиная с какого-то момента постоянна. Вводя естественное условие, что алгоритм соблюдает порядок обхода кластеров, получаем, что центры кластеров не будут меняться, если это не уменьшит целевой функции. Тогда и центры кластеров не меняются с какого-то момента.

```python
import numpy as np
import pylab as plt
'''
Performs the Principal Coponent analysis of the Matrix F
Matrix must be n * l dimensions
where n is # features
l is # samples
'''

def PCA(F, varRetained = 0.95, show = False):
    # Input
    #     F - initaial matrix 
    # Compute Covariance Matrix Sigma
    # Input
    (n, l) = F.shape
    # Compute eigenvectors and eigenvalues of Sigma by SVD
    # U, V - matrix, d - array: Sigma = U * np.diag(d) * V
    U, d, V = np.linalg.svd(F)
   
    # compute the value m: number of minumum features that retains the given variance varRetaine
    dTot = np.sum(d)
    var_i = np.array([np.sum(d[: i + 1]) / \
                dTot * 100.0 for i in range(n)])

    m = np.where(var_i > varRetained * 100)[0][0]
    print('%.2f %% variance retained in %d dimensions'%(varRetained, m)) # верните число m и точность, которая достигается при этом числе главных компонент

    # plot the variance plot
    if show:
        plt.plot(var_i)
        plt.xlabel('Number of Features')
        plt.ylabel(' Percentage Variance retained')
        plt.title('PCA $\% \sigma^2 $ vs # features')
        plt.show()

    # compute the reduced dimensional features by projection
    U_reduced = U[0:m,:]
    G = np.dot(U_reduced, F)

    return G, U_reduced
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
data, target = digits['data'], digits['target']
```

```python
from sklearn.manifold import TSNE
tsne_out = TSNE().fit_transform(data)
```

```python
from sklearn.cluster import DBSCAN
from tqdm import tqdm
from sklearn.metrics import adjusted_rand_score, adjusted_mutual_info_score, v_measure_score
```

```python
eps_range = np.linspace(0.1, 4.0)
metrics = {
            'AMI': [],
            'ARI': [],
            'V-score': []
    
          }
for eps in tqdm(eps_range):
    model = DBSCAN(eps=eps).fit(tsne_out)
    labels = model.labels_
    metrics['V-score'].append(v_measure_score(target, labels))
    metrics['AMI'].append(adjusted_mutual_info_score(target, labels))
    metrics['ARI'].append(adjusted_rand_score(target, labels))
from matplotlib import pyplot
%matplotlib inline
pyplot.figure(figsize=(16,10))
for metric in ['AMI', 'ARI', 'V-score']:
    pyplot.plot(eps_range, metrics[metric], label=metric)
pyplot.legend()
pyplot.title('Quality metrics vs eps')
```

Наилучшее качество достигается при $eps \in [1.0, 2.0]$

```python
n_samples_range = range(1,50)
pyplot.figure(figsize=(16,50))
for (index, eps) in enumerate(np.linspace(1.0, 2.0, 5)):
    metrics = {
            'AMI': [],
            'ARI': [],
            'V-score': []
    
          }
    for n_samples in tqdm(n_samples_range):
        model = DBSCAN(eps=eps, min_samples=n_samples).fit(tsne_out)
        labels = model.labels_
        metrics['V-score'].append(v_measure_score(target, labels))
        metrics['AMI'].append(adjusted_mutual_info_score(target, labels))
        metrics['ARI'].append(adjusted_rand_score(target, labels))
    pyplot.subplot(5, 1, index+1)
    for metric in ['AMI', 'ARI', 'V-score']:
        pyplot.plot(n_samples_range, metrics[metric], label=metric)
    pyplot.legend()
    pyplot.title('Quality metrics vs min_samples, eps = {}'.format(eps))
```
